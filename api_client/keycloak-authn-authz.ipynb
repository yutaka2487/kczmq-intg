```python
from keycloak import KeycloakOpenID
```

# Keyclaokでクライアント認証

```python
keycloak_openid = KeycloakOpenID(
    server_url="http://keycloak:8080/auth/",
    realm_name="kczmq-intg",
    client_id="client-proto",
    client_secret_key="AdbObscdCOvKyatYlcPpOEx4MKDf2Rhj",
)
keycloak_openid.well_known()
```

# アクセストークン取得

```python
token = keycloak_openid.token(grant_type=["client_credentials"])
token["access_token"]
```

# アクセストークンのデコード＋マッパー情報取得

```python
KEYCLOAK_PUBLIC_KEY = "-----BEGIN PUBLIC KEY-----\n" + keycloak_openid.public_key() + "\n-----END PUBLIC KEY-----"
options = {"verify_signature": True, "verify_aud": False, "verify_exp": True}
keycloak_openid.decode_token(token['access_token'], key=KEYCLOAK_PUBLIC_KEY, options=options)
```
